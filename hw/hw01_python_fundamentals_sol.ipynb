## Exercise 1: Gas Law Calculations (25 points)

The ideal gas law is given by:

$$PV = nRT$$

where:
- $P$ = pressure (atm)
- $V$ = volume (L)
- $n$ = number of moles (mol)
- $R$ = ideal gas constant = 0.08206 L¬∑atm/(mol¬∑K)
- $T$ = temperature (K)

### Part A: Define a Function (12 points)

Create a function called `calculate_pressure` that takes volume (L), moles (mol), and temperature (K) as inputs and returns the pressure in atm. Use proper formatting to print the result with 3 decimal places.

```python
# Import necessary libraries
R = 0.08206 # Ideal gas constant L¬∑atm/(mol¬∑K)

# Define your function here
# AI assistance: Yes - Generated the function definition and test code based on the problem description.
def calculate_pressure(volume_L, moles, temperature_K):
    """Calculates pressure using the ideal gas law (PV=nRT)."""
    pressure_atm = (moles * R * temperature_K) / volume_L
    return pressure_atm

# Test your function with: V = 10.0 L, n = 2.5 mol, T = 298.15 K
moles_test = 2.5
volume_test = 10.0
temperature_test = 298.15

calculated_pressure = calculate_pressure(volume_test, moles_test, temperature_test)

# Print the result in a formatted string
print(f"At {temperature_test} K, {moles_test} moles in {volume_test} L results in a pressure of {calculated_pressure:.3f} atm")
```

## Exercise 2: Heat Capacity and Energy Calculations (25 points)

The heat required to change the temperature of a substance is given by:

$$q = m \cdot C_p \cdot \Delta T$$

where:
- $q$ = heat energy (J)
- $m$ = mass (g)
- $C_p$ = specific heat capacity (J/(g¬∑K))
- $\Delta T$ = temperature change (K or ¬∞C)

### Part A: Energy Calculation Function (12 points)

Write a function `calculate_heat` that takes mass (g), specific heat (J/(g¬∑K)), initial temperature (¬∞C), and final temperature (¬∞C) as inputs and returns the heat energy in both Joules and kiloJoules. The function should print a formatted statement showing both values.

Test with: Heating 500 g of water (Cp = 4.184 J/(g¬∑K)) from 20¬∞C to 100¬∞C.

```python
# AI assistance: Yes - Generated the function definition and test code based on the problem description.

def calculate_heat(mass_g, specific_heat_J_gK, initial_temp_C, final_temp_C):
    """Calculates heat energy (q = m * Cp * ŒîT).

    Args:
        mass_g (float): Mass of the substance in grams.
        specific_heat_J_gK (float): Specific heat capacity in J/(g¬∑K).
        initial_temp_C (float): Initial temperature in Celsius.
        final_temp_C (float): Final temperature in Celsius.

    Returns:
        tuple: A tuple containing heat energy in Joules and kiloJoules.
    """
    delta_T = final_temp_C - initial_temp_C
    q_joules = mass_g * specific_heat_J_gK * delta_T
    q_kilojoules = q_joules / 1000

    print(f"Heat energy: {q_joules:.2f} J or {q_kilojoules:.3f} kJ")
    return q_joules, q_kilojoules

# Test case: Heating 500 g of water (Cp = 4.184 J/(g¬∑K)) from 20¬∞C to 100¬∞C.
mass = 500 # g
Cp_water = 4.184 # J/(g¬∑K)
initial_temp = 20 # ¬∞C
final_temp = 100 # ¬∞C

calculate_heat(mass, Cp_water, initial_temp, final_temp)
```

```python
# AI assistance: Yes - Generated the function definition, calculations, and error handling based on the problem description. Also included the test cases.

def dilution_calculator(C1=None, V1=None, C2=None, V2=None):
    """Calculates a missing variable (C1, V1, C2, or V2) in the dilution equation (C1V1 = C2V2).

    Args:
        C1 (float, optional): Initial concentration. Defaults to None.
        V1 (float, optional): Initial volume. Defaults to None.
        C2 (float, optional): Final concentration. Defaults to None.
        V2 (float, optional): Final volume. Defaults to None.

    Returns:
        float: The calculated missing value.

    Raises:
        ValueError: If more than one or zero variables are None.
    """
    known_vars = [var for var in [C1, V1, C2, V2] if var is not None]
    none_count = 4 - len(known_vars)

    if none_count == 0:
        raise ValueError("No variable is None. Please provide exactly three known values and one None value.")
    elif none_count > 1:
        raise ValueError("More than one variable is None. Please provide exactly three known values and one None value.")

    if C1 is None:
        calculated_value = (C2 * V2) / V1
        unit = "M"
        print(f"Calculated C1: {calculated_value:.3f} {unit}")
    elif V1 is None:
        calculated_value = (C2 * V2) / C1
        unit = "L" # Assuming input volumes are in L, output will be in L
        print(f"Calculated V1: {calculated_value:.3f} {unit}")
    elif C2 is None:
        calculated_value = (C1 * V1) / V2
        unit = "M"
        print(f"Calculated C2: {calculated_value:.3f} {unit}")
    elif V2 is None:
        calculated_value = (C1 * V1) / C2
        unit = "L" # Assuming input volumes are in L, output will be in L
        print(f"Calculated V2: {calculated_value:.3f} {unit}")
    else:
        # This case should ideally be caught by none_count == 0, but as a safeguard
        raise ValueError("Unexpected error: No None variable found despite passing the initial check.")

    return calculated_value

# Test cases:
# 1. If you have 100 mL of 0.5 M NaCl and dilute to 250 mL, what is the final concentration?
print("\nTest Case 1: Calculate C2")
final_concentration = dilution_calculator(C1=0.5, V1=100, C2=None, V2=250)

# 2. What volume of 2.0 M HCl do you need to make 500 mL of 0.1 M HCl?
print("\nTest Case 2: Calculate V1")
initial_volume = dilution_calculator(C1=2.0, V1=None, C2=0.1, V2=500)
```

### Part B: Serial Dilution Analysis (6 points)

A serial dilution involves repeatedly diluting a solution. Starting with 10 mL of 1.0 M solution:
1. Dilute to 50 mL (dilution 1)
2. Take 10 mL of dilution 1 and dilute to 50 mL (dilution 2)
3. Repeat this process 5 times total

Create a numpy array showing the concentration after each dilution and print the results in a formatted table.

```python
# AI assistance: Yes - Generated code to perform serial dilution calculations and present results in a formatted table.

import numpy as np

# Initial conditions
initial_concentration = 1.0  # M
initial_volume_aliquot = 10  # mL
final_volume_dilution = 50   # mL
num_dilutions = 5

# Calculate the dilution factor for each step
dilution_factor = final_volume_dilution / initial_volume_aliquot

# Array to store concentrations after each dilution
concentrations = np.zeros(num_dilutions)
current_concentration = initial_concentration

print("Serial Dilution Analysis")
print("--------------------------------")
print("Dilution # | Concentration (M)")
print("--------------------------------")

for i in range(num_dilutions):
    current_concentration = current_concentration / dilution_factor
    concentrations[i] = current_concentration
    print(f"{i+1:10d} | {concentrations[i]:17.5f}")

# Optional: Print the final concentrations array
# print("\nConcentrations array:", concentrations)
```

## Exercise 4: Reaction Rate Analysis (25 points)

The rate of a first-order chemical reaction depends on concentration:

$$r = k \cdot C$$

4. Determine at what time the concentration drops to 25% of its initial value

```python
# AI assistance: Yes - Generated the function definition, calculations, and formatted output for the time-dependent concentration and the time to 25% concentration.

import numpy as np

def concentration(t, C0, k):
    """Calculates the concentration C(t) for a first-order reaction.

    Args:
        t (float): Time in seconds.
        C0 (float): Initial concentration (mol/L).
        k (float): Rate constant (1/s).

    Returns:
        float: Concentration at time t (mol/L).
    """
    return C0 * np.exp(-k * t)

# Given values
C0 = 2.0  # M
k = 0.05  # s‚Åª¬π

# Times to calculate concentrations at
times = np.array([0, 10, 20, 30, 40, 50])

# Calculate concentrations at these times
concentrations = np.array([concentration(t, C0, k) for t in times])

# Store results in a formatted table
print("Time (s) | Concentration (M)")
print("---------------------------------")
for i in range(len(times)):
    print(f"{times[i]:8d} | {concentrations[i]:17.3f}")

# Determine at what time the concentration drops to 25% of its initial value
# 0.25 * C0 = C0 * exp(-k*t)
# 0.25 = exp(-k*t)
# ln(0.25) = -k*t
# t = -ln(0.25) / k

t_25 = -np.log(0.25) / k
print(f"\nConcentration drops to 25% of initial value at t = {t_25:.2f} seconds")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# AI assistance: Yes - Generated code to create a single plot showing concentration vs time for three different rate constants, including labels, title, and legend.

# Given values
C0 = 2.0  # Initial concentration (M)
times_plot = np.linspace(0, 100, 200) # Time from 0 to 100 seconds, 200 points for smooth curve

# Different rate constants
k_slow = 0.01 # s‚Åª¬π
k_medium = 0.05 # s‚Åª¬π
k_fast = 0.10 # s‚Åª¬π

# Calculate concentrations for each rate constant
conc_slow = concentration(times_plot, C0, k_slow)
conc_medium = concentration(times_plot, C0, k_medium)
conc_fast = concentration(times_plot, C0, k_fast)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(times_plot, conc_slow, label=f'k = {k_slow} s‚Åª¬π (Slow)', color='blue', linestyle='-')
plt.plot(times_plot, conc_medium, label=f'k = {k_medium} s‚Åª¬π (Medium)', color='green', linestyle='--')
plt.plot(times_plot, conc_fast, label=f'k = {k_fast} s‚Åª¬π (Fast)', color='red', linestyle=':')

# Labeling and title
plt.xlabel('Time (s)')
plt.ylabel('Concentration (M)')
plt.title('Concentration vs. Time for First-Order Reaction at Different Rate Constants')
plt.legend()
plt.grid(True)
plt.show()
```
